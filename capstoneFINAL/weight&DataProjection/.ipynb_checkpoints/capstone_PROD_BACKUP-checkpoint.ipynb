# Database

## Setup

```python
# establish the connection (conn) and instantiate the DB
conn = sqlite3.connect("ExercisePortolio.db")
```

```python
# create a 'cursor' object to execute SQL statements
cur = conn.cursor()
```

***

## CRUD: Create

```python
# re-read in to get all rows and columns
final_df = pd.read_csv(file_location+file_name)
```

```python
# create the table 'Master_Dataset' in the ExercisePortfolio DB
final_df.to_sql("Master_Dataset", conn, if_exists="replace", index=False)
```

```python
# show the existing tables
pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
```

***

## CRUD: Read

```python
# read the top 5 records from the Master_Dataset table
pd.read_sql("""SELECT * FROM Master_Dataset LIMIT 5""", conn)
```

```python
# constrain to only where Projectible
pd.read_sql("""SELECT * FROM Master_Dataset WHERE Projectible='Yes'""", conn)
```

***

## CRUD: Update

```python
# create the column on the table
cur.execute("""ALTER TABLE Master_Dataset ADD COLUMN Projected1RM FLOAT""")
```

```python
# show that the 'shell' column has been created - but not yet populated (all values = None)
pd.read_sql("SELECT * FROM Master_Dataset", conn)
```

```python
# add the projects column to the pandas dataframe
final_df["Projected1RM"] = ea.get1RM(final_df.MaxWeightSet, final_df.Reps)
```

```python
# create new table on the database to join with Master_Dataset
join_columns = ["Username", "Exercise", "Projected1RM"]
final_df[join_columns].to_sql("projections", conn, if_exists="replace", index=False)
```

```python
# print out the top 10 records from the newly created table
pd.read_sql("SELECT * FROM projections LIMIT 10", conn)
```

```python
# update Master_Dataset by joining the new column by Username and Exercise
sql_query= """UPDATE Master_Dataset
              SET Projected1RM =
              (SELECT Projected1RM
                   FROM projections
                   WHERE projections.Username = Master_Dataset.Username
                       AND projections.Exercise = Master_Dataset.Exercise)"""

conn.execute(sql_query)
```

```python
# Projected1RM has been successfully updated on Master_Dataset
pd.read_sql("SELECT Exercise, BodyPart, Projectible, Reps, Sets, MaxWeightSet, CurrentPR, GoalPR, ProjPR, Projected1RM FROM Master_Dataset", conn)
```

***

## CRUD: Delete

```python
# what are the existing tables
pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
```

```python
# drop the temp table 'projections'
cur.execute("DROP TABLE IF EXISTS projections")
```

```python
# now list which tables exist - show that projections was deleted
cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
print(cur.fetchall())
```
